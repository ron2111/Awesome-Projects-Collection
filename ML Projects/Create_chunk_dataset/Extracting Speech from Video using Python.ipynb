```python
dir_path='/content/drive/MyDrive/SPEECH DATA/extract Audio from video/videos'
save_dir='/content/drive/MyDrive/SPEECH DATA/extract Audio from video/converted'

import os,sys
for path, subdirs, files in os.walk(dir_path):
  for name in files:
    if name.endswith('.mp4'):
      clip_path=dir_path+'/' + name

      model_id, img_id= name.rsplit('.',1)
      # print(model_id)
  
      clip = mp.VideoFileClip(clip_path)

      save_file_name= save_dir + '/'+ 'converted'+model_id +'.wav'
      clip.audio.write_audiofile(save_file_name)
```
